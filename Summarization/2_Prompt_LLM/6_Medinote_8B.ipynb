# Import model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "AGBonnet/medinote-7b"

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Use FP16 (recommended on A100)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype="auto"  # or torch.float16
)
```

# Inference

```python
from transformers import AutoConfig

config = AutoConfig.from_pretrained("AGBonnet/medinote-7b")
print(config.max_position_embeddings)
```

```python
inputs = tokenizer(dialogue, return_tensors="pt").to("cuda")
input_token_length = inputs["input_ids"].shape[-1]

outputs = model.generate(**inputs)
# Extract newly generated tokens (excluding input prompt)
generated_tokens = outputs[0][input_token_length:]
# Decode tokens to text
generated_text = tokenizer.decode(generated_tokens, skip_special_tokens=True)

print(generated_text)
```
